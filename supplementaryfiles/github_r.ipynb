## Count references to GitHub.com in PubMed Central

```r
# Code by Yasset Perez-Riverol
# Load libraries
require(devtools)
require(rebi)

# Function to count the
counts = function(year) {
  query = sprintf("github.com AND PUB_YEAR:%d", year)
  print(query)
  my_data <- epmc_search(query = query)
  my_data$hit_count
}

# Define the period
years = 2009:2015

# Count the number of hits over the period
hits = sapply(years, counts)
```

```r
# Plot references to github.com over time
plot(years, hits, log='', pch=19, bty='n',
     xlab='year of publication',
     ylim=c(0, 2000),
     las=1, main='github.com references on Pubmed Central')
```
